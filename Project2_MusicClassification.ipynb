# y-label 정리

### genre를 숫자로 label 바꾸기

```python
n = 0
for genre in train_df["genre_top"].unique():
    for index, row in train_df.iterrows():
        if row["genre_top"] == genre:
            train_df.loc[index, 'genre_top'] = n
    n += 1
    
train_df
```

### genre를 one-hot 코딩으로 하기 위해 label 변환

```python
for index, row in train_df.iterrows():
    if row["genre_top"] == 0:
            row['genre_top'] = [np.array((1,0,0,0,0,0,0,0))]
    if row["genre_top"] == 1:
            row['genre_top'] = [np.array((0,1,0,0,0,0,0,0))]
    if row["genre_top"] == 2:
            row['genre_top'] = [np.array((0,0,1,0,0,0,0,0))]
    if row["genre_top"] == 3:
            row['genre_top'] = [np.array((0,0,0,1,0,0,0,0))]
    if row["genre_top"] == 4:
            row['genre_top'] = [np.array((0,0,0,0,1,0,0,0))]
    if row["genre_top"] == 5:
            row['genre_top'] = [np.array((0,0,0,0,0,1,0,0))]
    if row["genre_top"] == 6:
            row['genre_top'] = [np.array((0,0,0,0,0,0,1,0))]
    if row["genre_top"] == 7:
            row['genre_top'] = [np.array((0,0,0,0,0,0,0,1))]

    
train_df
```
